Importing the Libraries Required

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis


```python
# loading the diabetes dataset to pandas dataframe
diabetes_set = pd.read_csv('/content/diabetes.csv')
```

```python
# printing 1st five rows of dataset
diabetes_set.head(20)
```

```python
# Number of rows and columns from this dataset table
diabetes_set.shape
```

```python
# getting the statistical measures
diabetes_set.describe()
```

```python
diabetes_set['Outcome'].value_counts()
```

0 -> Non Diabetic

1 -> Diabetic

```python
diabetes_set.groupby('Outcome').mean()
```

```python
# Separating data and label
x=diabetes_set.drop(columns='Outcome',axis=1)
y=diabetes_set['Outcome']
print(x)
print(y)
```

Data Standardization


```python
scaler=StandardScaler()
scaler.fit(x)
sdata=scaler.transform(x)
print(sdata)
import joblib
# Save the fitted scaler
joblib.dump(scaler, 'diabetes_scaler.pkl')
```

```python
x=sdata
y=diabetes_set['Outcome']
print(x)
print(y)
```

Train Test split

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
print(x.shape,xtrain.shape,xtest.shape)#20% test 80% train
```

Training the Model

```python
classifier =svm.SVC(kernel='linear')
#training the support vector machine classifier
classifier.fit(xtrain,ytrain)
```

Model Evaluation


accuracy score

```python
#Accuracy score on Training data
xtrain_acc_prediction=classifier.predict(xtrain)
training_data_acc=accuracy_score(xtrain_acc_prediction,ytrain)
```

```python
print('Accuracy Score of training data:',training_data_acc)
```

```python
#Accuracy score on Test data
xtest_acc_prediction=classifier.predict(xtest)
test_data_acc=accuracy_score(xtest_acc_prediction,ytest)
print('Accuracy Score of test data:',test_data_acc)
```

```python
# Precision, Recall, and F1-Score for Test Data
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(ytest, xtest_acc_prediction, average='weighted')  # 'weighted' accounts for class imbalance
recall = recall_score(ytest, xtest_acc_prediction, average='weighted')
f1 = f1_score(ytest, xtest_acc_prediction, average='weighted')

# Print the metrics
print(f'Precision on Test Data: {precision:.4f}')
print(f'Recall on Test Data: {recall:.4f}')
print(f'F1 Score on Test Data: {f1:.4f}')
```

Making a Prediction System

```python
input_data=(6	,148,	72,	35,	0,	33.6,	0.627,	50)
#convert input dat to numpy array
ind=np.asarray(input_data)
#reshape of array as we re predicting for one instance
ind_reshape=ind.reshape(1,-1)
#Standardize the input data to work
std_data=scaler.transform(ind_reshape)
print(std_data)
#prediction
prediction=classifier.predict(std_data)
print(prediction)
if(prediction[0]==0):
  print('The Person is Non-Diabetic')
else:
  print('The Person is Diabetic')
```
